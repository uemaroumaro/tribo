```python
import pandas as pd
case_df_Tcluster_sec = pd.read_csv("C:/tmp/Treport/caseframe_Tcluster_sec.csv", encoding='shift-jis')
```

```python
import random
N = 1000
N = len(case_df_Tcluster_sec[u"報告書_id"].drop_duplicates())
ExNlist = random.sample(case_df_Tcluster_sec[u"報告書_id"].drop_duplicates(), N)
case_df_tmp = case_df_Tcluster_sec.ix[case_df_Tcluster_sec[u"報告書_id"].map(lambda x: x in ExNlist), :]
```

```python
import re
cases = case_df_tmp[u"主体"] + " " + case_df_tmp[u"起点"] + " " + case_df_tmp[u"対象"] + " " + case_df_tmp[u"状況"] + " " + \
                 case_df_tmp[u"着点"] + " " + case_df_tmp[u"手段"] + " " + case_df_tmp[u"関係"] + " " + case_df_tmp[u"動詞"]
for i in range(0, len(cases)):
    cases[i] = re.sub(r" +", u" ", cases[i].strip())
pro_zeroNoun=case_df_tmp.ix[case_df_tmp[u"事象"]!=cases,:]        
pro_zeroNoun
```

```python
from Language import Language
pronoun_id=[]
for Report_id in case_df_tmp[u"報告書_id"].drop_duplicates():
    print Report_id
    case_df_perR = case_df_tmp[case_df_tmp[u"報告書_id"]==Report_id]
    for first_Sen, Sentence_id in enumerate(case_df_perR[u"文_id"].drop_duplicates()):
        
        for line in case_df_perR[case_df_tmp[u"文_id"]==Sentence_id].iterrows():
            #代名詞の補完
            for di, l in enumerate(line[1][4:11].values):
                if l != u" ":
                    Lan = Language(l)
                    outList = Lan.getMorpheme()
                    if set([u"代名詞"]).intersection(set([outList[i][2] for i in range(len(outList))])) :     
                        pronoun_id.append((Report_id,Sentence_id,line[1][2]))
                        break
```

```python
case_df_tmp[u"proNoun_id"]=[(r,s,v) for r,s,v in zip(case_df_tmp[u"報告書_id"],case_df_tmp[u"文_id"],case_df_tmp[u"動詞_id"])]
case_df_tmp.ix[case_df_tmp[u"proNoun_id"].map(lambda x: x in pronoun_id), :]
```

```python
import pandas as pd
case_df_Tcluster_eval = pd.read_csv(u"D:/tmp/Treport/caseframe_cluster/Jaccard_0.4/Evaluation/case_frame_クラスター-2.csv", encoding='shift-jis')
```

```python
case_df_Tcluster_eval=case_df_Tcluster_eval[case_df_Tcluster_eval[u"sec_change"].notnull()]
Precisions=[]
Precisions_continue=[]
Precisions_change=[]

for Report_id in case_df_Tcluster_eval[u"報告書_id"].drop_duplicates():
    case_df_perR = case_df_Tcluster_eval[case_df_Tcluster_eval[u"報告書_id"]==Report_id]
    for first_Sen, (Record_id, sec_change) in enumerate(zip(case_df_perR[u"レコード_id"], case_df_perR[u"sec_change"])):
        if first_Sen==0:
            Record_id_pre=Record_id
            continue
        
        if sec_change==0:
            if Record_id_pre==Record_id:
                Precisions.append(1)
                Precisions_continue.append(1)
            elif Record_id_pre!=Record_id:
                Precisions.append(0)
                Precisions_continue.append(0)                
        elif sec_change==1:
            if Record_id_pre==Record_id:
                Precisions.append(0)
                Precisions_change.append(0)
            elif Record_id_pre!=Record_id:
                Precisions.append(1)
                Precisions_change.append(1)                
        Record_id_pre=Record_id
```

```python
print round(float(Precisions.count(1))/len(Precisions), 3)
```

```python
print round(float(Precisions_continue.count(1))/len(Precisions_continue), 3)
print round(float(Precisions_change.count(1))/len(Precisions_change), 3)
```
