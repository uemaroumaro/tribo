```python
import pandas as pd
case_df_Tcluster_eval = pd.read_csv(u"D:/tmp/Treport/caseframe_cluster/Jaccard_0.4/Evaluation/case_frame_クラスター-2.csv", encoding='shift-jis')
```

```python
case_df_Tcluster_eval=case_df_Tcluster_eval[case_df_Tcluster_eval[u"sec_change"].notnull()]
Precisions=[]
Precisions_continue=[]
Precisions_change=[]

for Report_id in case_df_Tcluster_eval[u"報告書_id"].drop_duplicates():
    case_df_perR = case_df_Tcluster_eval[case_df_Tcluster_eval[u"報告書_id"]==Report_id]
    for first_Sen, (Record_id, sec_change) in enumerate(zip(case_df_perR[u"レコード_id"], case_df_perR[u"sec_change"])):
        if first_Sen==0:
            Record_id_pre=Record_id
            continue
        
        if sec_change==0:
            if Record_id_pre==Record_id:
                Precisions.append(1)
                Precisions_continue.append(1)
            elif Record_id_pre!=Record_id:
                Precisions.append(0)
                Precisions_continue.append(0)                
        elif sec_change==1:
            if Record_id_pre==Record_id:
                Precisions.append(0)
                Precisions_change.append(0)
            elif Record_id_pre!=Record_id:
                Precisions.append(1)
                Precisions_change.append(1)                
        Record_id_pre=Record_id
```

```python
print u"all:", round(float(Precisions.count(1))/len(Precisions), 3)
print u"continue:", round(float(Precisions_continue.count(1))/len(Precisions_continue), 3)
print u"change:", round(float(Precisions_change.count(1))/len(Precisions_change), 3)
```

```python
Record_id = dict()#文_id:レコード_id
Record_id[(case_df_tmp.ix[0,:][u"報告書_id"],case_df_tmp.ix[0,:][u"文_id"])] = 0
tail_key=-1
for Report_id in case_df_tmp[u"報告書_id"].drop_duplicates():
    Noun_pre = dict()
    print Report_id
    case_df_perR = case_df_tmp[case_df_tmp[u"報告書_id"]==Report_id]
    for first_Sen, Sentence_id in enumerate(case_df_perR[u"文_id"].drop_duplicates()):
        
        for line in case_df_perR[case_df_tmp[u"文_id"]==Sentence_id].iterrows():
            #print line[1][1] 
            #print line[1][3]
            if line[1][1] not in Noun_pre.keys():
                Noun_pre[line[1][1]]=[l for l in line[1][4:11].values if l!=u" "]
            else:
                Noun_pre[line[1][1]]= Noun_pre[line[1][1]]+[l for l in line[1][4:11].values if l!=u" "]
        
        #前の文に含まれる名詞が含まれているか        
        if first_Sen==0 and tail_key!=-1:
            Record_id[(Report_id, Sentence_id)] = Record_id[tail_key]+1
            continue
        for pre_i in range(3, 0, -1):
            if line[1][1]-pre_i in Noun_pre.keys():
                if set(Noun_pre[line[1][1]]).intersection(set(Noun_pre[line[1][1]-pre_i])):
                    for pre_j in range(pre_i,-1,-1):
                        if line[1][1]-pre_j in Noun_pre.keys():
                            Record_id[(Report_id, Sentence_id-pre_j)] = Record_id[(Report_id, line[1][1]-pre_i)]
                    break
                else:
                    Record_id[(Report_id, Sentence_id)] = Record_id[(Report_id, line[1][1]-pre_i)]+1 
        while (Report_id, Sentence_id) not in Record_id.keys():
            pre_i+=1
            if (Report_id, Sentence_id-pre_i) in Record_id.keys():
                Record_id[(Report_id, Sentence_id)] = Record_id[(Report_id, line[1][1]-pre_i)]+1
        if first_Sen ==len(case_df_perR[u"文_id"].drop_duplicates())-1:
            tail_key = (Report_id, Sentence_id)
        

case_df_tmp[u"レコード_id"] = [(i,j) for i,j in zip(case_df_tmp[u"報告書_id"], case_df_tmp[u"文_id"])]
case_df_tmp[u"レコード_id"] = case_df_tmp[u"レコード_id"].map(lambda x: Record_id[x])  
```
